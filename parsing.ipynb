Check if the sentence " The cat sat on the mat" is syntactically correct using python.

```python
import nltk
from nltk import CFG

# Step 1: Define grammar (all lowercase!)
grammar = CFG.fromstring("""
S -> NP VP
NP -> Det N
VP -> V PP
PP -> P NP
Det -> 'the'
N -> 'cat' | 'mat'
V -> 'sat'
P -> 'on'
""")

# Step 2: Input sentence (also lowercase!)
sentence = "the cat sat on the mat".split()

# Step 3: Parse
parser = nltk.ChartParser(grammar)

trees = list(parser.parse(sentence))

# Step 4: Output result
if trees:
    print("✔️ Sentence is syntactically correct!")
    for tree in trees:
        tree.pretty_print()
else:
    print("❌ Sentence is NOT syntactically correct.")

```
